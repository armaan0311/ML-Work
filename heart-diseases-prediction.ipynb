```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from fastai.tabular.all import *
import plotly.express as px
```

```python
df = pd.read_csv('/kaggle/input/heart-disease-prediction/Heart_Disease_Prediction.csv')
df
```

```python
 labels=["Healthy","Heart Disease"]

healthy_or_not = df['Heart Disease'].value_counts().tolist()
values = [healthy_or_not[0], healthy_or_not[1]]

fig = px.pie(values=df['Heart Disease'].value_counts(), names=labels , width=700, height=400, color_discrete_sequence=["#F7C0BB","#8690FF"]
             ,title="Healthy / Heart Disease")
fig.show()
```

```python
dls = TabularDataLoaders.from_df(df, y_names="Heart Disease",
                                 cat_names=['Sex',
                                            'Chest pain type',
                                            'FBS over 120',
                                            'EKG results',
                                            'Exercise angina',
                                            'Slope of ST',
                                            'Number of vessels fluro',
                                            'Thallium',
                                            'ST depression'],
                                 cont_names=['Cholesterol',
                                             'BP',
                                             'Age',
                                             'Max HR'],
                                 procs=[Categorify,
                                        FillMissing,
                                        Normalize],
                                 bs=32,
                                 device=torch.device('cpu'))
```

```python
learn = tabular_learner(dls, metrics=accuracy)
learn.fit_one_cycle(15, lr_max=slice(3e-1, 1e-1))
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.print_classification_report()
```
